```python
import numpy as np
import matplotlib.pyplot as plt

N = 8 #Tamanho da FFT

Xkm = []
x = [1, 1, 1, 1, 1, 1, -1, -1]

fig, ax = plt.subplots(1,3, figsize=(20,4))
ax[0].stem(x,use_line_collection=True,)
ax[0].set_title("Signal",fontsize=16)

Xk = np.fft.fft(x,N)
ax[1].stem(Xk,use_line_collection=True,)
ax[1].set_title("Signal DFT",fontsize=16)

m = 1 #Deslocamento

for k in range(0,N):
    Xkm.append(Xk[k]*np.exp(-1j*2*np.pi*k*m/N))
 
xhat = np.fft.ifft(Xkm,N)

ax[2].stem(np.real(xhat),use_line_collection=True,)
ax[2].set_title("Inverse DFT",fontsize=16)
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 8 #Tamanho da FFT

x = [1, 1, 1, 1, 1, 1, 1, 1]
h = [0.5, 0.5]

fig, ax = plt.subplots(2,2, figsize=(18,10))
ax[0][0].stem(x,use_line_collection=True,)
ax[0][0].set_title("Signal x",fontsize=16)

ax[0][1].stem(h,use_line_collection=True,)
ax[0][1].set_title("Signal h",fontsize=16)

Xk = np.fft.fft(x,N)
Hk = np.fft.fft(h, N)

print(Xk)
print(Hk)
Yk = []

for i in range(0,N):
    Yk.append(Xk[i]*Hk[i])
print(Yk)
    
y = np.fft.ifft(Yk,N)

ax[1][0].stem(np.real(y),use_line_collection=True,)
ax[1][0].set_title("Inverse DFT",fontsize=16)

y_linear = np.convolve(x, h)

ax[1][1].stem(np.real(y_linear),use_line_collection=True,)
ax[1][1].set_title("Linear Convolution",fontsize=16)
```
